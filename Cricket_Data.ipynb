```python
import numpy as np
import pandas as pd
import json
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## Match Summary

```python
with open('ODI_wc_match_results.json') as f:
  data = json.load(f)
```

```python
df_match = pd.DataFrame(data['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match = df_match.rename(columns={"scorecard": "match_id"})
df_match.head()
```

## Batting Summary

```python
with open('ODI_wc_batting_summary.json') as f:
  data = json.load(f)
```

```python
records = []
for rec in data:
  records.extend(rec['battingSummary'])

print(len(records))
```

```python
records[:5]
```

```python
df_batting = pd.DataFrame(records)
df_batting.head(15)
```

```python
df_batting.isnull().sum()
```

```python
df_batting["out/not_out"] = np.where(df_batting["dismissal"].isnull(),"not_out","out")
df_batting.head(15)
```

```python
df_batting = df_batting.drop(columns=["dismissal"])
df_batting.isnull().sum()
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x: x.replace('†',''))
df_batting.head()
```

Now We need to connect both the tables using match_ids. For this we will make a dictionary using df_match as follows:

{
  
  team1 + "Vs" + team2 : match_id
  
  team2 + "Vs" + team1 : match_id

}

We dont know the order order of teams in match columns of df_batting.

**Warning :**
We need to mannualy assign the ids of Final and Semifinal matchs as they are played twice amongst the same teams

```python
match_ids_dict = {}

for idx, row in df_match.iterrows():
  key_1 = row["team1"] + " Vs " + row["team2"]
  key_2 = row["team2"] + " Vs " + row["team1"]

  match_ids_dict[key_1] = row["match_id"]
  match_ids_dict[key_2] = row["match_id"]

match_ids_dict
```

```python
print(df_batting["match"].unique())
```

```python
len(df_batting["match"].unique())
```

This is done manually looking at the match

```python
match_ids_dict['India Vs Australia'] = "ODI # 4705"
match_ids_dict['Australia Vs South Africa'] = "ODI # 4704"
match_ids_dict['India Vs New Zealand'] = "ODI # 4703"
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```

## Bowling Summary

```python
records = []
for rec in data:
  records.extend(rec['bowlingSummary'])

print(len(records))
```
